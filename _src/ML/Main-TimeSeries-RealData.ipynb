```python
import csv
import pandas
#from pandas.plotting import scatter_matrix
import numpy as np
#from scipy.stats import norm #for displaying normal dist
#import random #for generating random colors for plots
#from scipy.interpolate import spline #for smooth lineplots

class Traffic:
    def __init__(self):
        self.file = "_data/ShortCountsManhattan.csv"
        self.col_names = ["Time","W 30TH ST","W 38TH ST","W 36TH ST","E 55TH ST","E 48TH ST","E 48TH ST","E 40TH ST","W 34TH ST","W 34TH ST","E 57TH ST","W 30TH ST","W 38TH ST","W 40TH ST","E 41ST ST","W 46TH ST","E 54TH ST","E 40TH ST","E 50TH ST","W 26TH ST","W 44TH ST","W 38TH ST","E 60TH ST","W 30TH ST","E 36TH ST","E 38TH ST","W 28TH ST","E 40TH ST","E 44TH ST","W 30TH ST","W 38TH ST","W 36TH ST","W 44TH ST","E 57TH ST","E 58TH ST","E 42ND ST","E 57TH ST","W 42ND ST","W 34TH ST]"]
        self.trafficDataFrame = pandas.read_csv(self.file, header=0)
        #self.trafficDataFrame['fact_incident'] = "Normal"
        #print(self.trafficDataFrame)
        self.numpyTraffic = np.array(self.trafficDataFrame)
        self.noCols = len(self.numpyTraffic[0])
        self.noRows = len(self.numpyTraffic)
        print(self.noRows, self.noCols)
        

    def print_dataset(self, rows=1, col=100):

        if col==100:
            for i in range(rows):
                print()
                for j in range(0,self.noCols,1):
                    print(self.numpyTraffic[i][j], end=" ")
        else:
            for i in range(rows):
                print(self.numpyTraffic[i][col], end =" ")

    def add_factor(self, factor="fact_incident", reason="a", year=0, month=0, day=0):
        for eachRecord in self.numpyTraffic:
            if day == 0:
                if int(eachRecord[0][6:10]) == year and int(eachRecord[0][1:2]) == month:
                    eachRecord[7] = reason
                    
        #for eachRecord in self.numpyTraffic:
        #    if int(eachRecord[0][6:10]) == year:
        #        print(eachRecord)
    

    def plot_traffic(self, col=100):
        
        roads = ["W 30TH ST","W 38TH ST","W 36TH ST","E 55TH ST","E 48TH ST","E 48TH ST","E 40TH ST","W 34TH ST","W 34TH ST","E 57TH ST","W 30TH ST","W 38TH ST","W 40TH ST","E 41ST ST","W 46TH ST","E 54TH ST","E 40TH ST","E 50TH ST","W 26TH ST","W 44TH ST","W 38TH ST","E 60TH ST","W 30TH ST","E 36TH ST","E 38TH ST","W 28TH ST","E 40TH ST","E 44TH ST","W 30TH ST","W 38TH ST","W 36TH ST","W 44TH ST","E 57TH ST","E 58TH ST","E 42ND ST","E 57TH ST","W 42ND ST","W 34TH ST]"]
        import matplotlib.pyplot as traffic_plot
        import datetime

        # line cyclers adapted to colourblind people
        from cycler import cycler
        line_cycler   = (cycler(color=["#E69F00", "#56B4E9", "#009E73", "#0072B2", "#D55E00", "#CC79A7", "#F0E442"]) +
                        cycler(linestyle=["-", "--", "-.", ":", "-", "--", "-."]))
        marker_cycler = (cycler(color=["#E69F00", "#56B4E9", "#009E73", "#0072B2", "#D55E00", "#CC79A7", "#F0E442"]) +
                        cycler(linestyle=["none", "none", "none", "none", "none", "none", "none"]) +
                        cycler(marker=["4", "2", "3", "1", "+", "x", "."]))
        traffic_plot.rc("axes", prop_cycle=line_cycler)
        traffic_plot.figure()
        #traffic_plot.rc("font", family="serif", size=12.)
        traffic_plot.rc("savefig", dpi=200)
        traffic_plot.rc("legend", loc="best", fontsize="medium", fancybox=True, framealpha=0.5)
        traffic_plot.rc("lines", linewidth=2.5, markersize=10, markeredgewidth=2.5)

        anotate_text=[]
        x = []
        y = []
        for i in range (5):
            for eachRecord in self.numpyTraffic:
                
                #print(eachRecord)
                x.append(eachRecord[0])
                y.append(eachRecord[col+i])
            
                anotate_text.append(eachRecord[col+i])
            traffic_plot.plot(x, y, label=roads[col+i])
            x.clear()
            y.clear()

        print("numpyTraffic", self.numpyTraffic)

        """
        bbox = {'fc': '0.8', 'pad': 1}
        props = {'ha': 'center', 'va': 'bottom', 'bbox': bbox}
        for i, txt in enumerate(anotate_text):
            if txt != "Normal":
                traffic_plot.text(data[i][0], data[i][2]+1, txt, props, rotation=90) #data[i][2]
        """ 
       
        #traffic_plot.plot(X[0], Y[0])
        #plt.xticks(self.col_names)
        traffic_plot.xticks(rotation=90)
        #plt.yticks(np.arange(2400, 12000, step=1000))
        traffic_plot.xlabel("Time of day", fontsize=14)
        traffic_plot.ylabel("Number of cars", fontsize=14)
        #traffic_plot.title(roads[col])
        traffic_plot.legend()
        traffic_plot.savefig("plots/cars-roads-"+str(datetime.datetime.now())+".png", bbox_inches="tight")
        traffic_plot.show()
    
    def __pre_process_data(self, window=None, col=100):
        from sklearn.preprocessing import MinMaxScaler
        #from keras.wrappers.scikit_learn import KerasRegressor
        import time
        import datetime as dt
        from tempfile import TemporaryFile
        
        outfile = TemporaryFile()
        d = np.array(self.numpyTraffic)
        np.save("plots/ShortCountsManhattan", d)
        da =[]
        for i in range(1,len(self.numpyTraffic[0])):
            da.append(d[:,i])
        
        #print("da", da)
        #print("numpyTraffic", self.numpyTraffic)

        train = da[:int(len(da)*0.8)]
        test = da[int(len(da)*0.8):]

        #print("train\n",train)
        #print("test\n",test)
        
        training_set = []
        for i in range(len(train)):
            training_set.append(int(train[i][1]))

        #print("training_set\n",np.array(training_set).reshape(-1,1))

        sc = MinMaxScaler(feature_range = (0, 1))
        training_set_scaled = sc.fit_transform(np.array(training_set).reshape(-1,1))
        #print("training_set_scaled\n",training_set_scaled)

        X_train = []
        y_train = []
        print("Data len:",len(da))
        print("Training data len:",len(training_set_scaled))
        print("Testing data len:",len(test))
        for i in range(window,len(train)-1):
            X_train.append(training_set_scaled[i-window:i,0])
            y_train.append(training_set_scaled[i,0])
        
        X_train, y_train = np.array(X_train), np.array(y_train)

        X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

        return sc, X_train, y_train, test, training_set

    def __build_regressor(self, X_train, y_train, units=50, dropout=0.2):
        from keras.models import Sequential
        from keras.layers import Dense
        from keras.layers import LSTM
        from keras.layers import Dropout

        #print(X_train)

        # Initialising the RNN
        regressor = Sequential()
        # Adding the first LSTM layer and some Dropout regularisation
        regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
        regressor.add(Dropout(0.2))

        # Adding a second LSTM layer and some Dropout regularisation
        regressor.add(LSTM(units = 50, return_sequences = True))
        regressor.add(Dropout(0.2))

        # Adding a third LSTM layer and some Dropout regularisation
        regressor.add(LSTM(units = 50, return_sequences = True))
        regressor.add(Dropout(0.2))

        # Adding a fourth LSTM layer and some Dropout regularisation
        regressor.add(LSTM(units = 50))
        regressor.add(Dropout(0.2))

        # Adding the output layer
        regressor.add(Dense(units = 1))

        # Compiling the RNN
        optim='adam'
        #optim  = 'Adagrad' 
        regressor.compile(optimizer = optim, loss = 'mean_squared_error', metrics=['mae','accuracy'])

        return regressor

    def __fit(self, regressor, X_train, y_train, epochs=10, batch_size=32):
        # Fitting the RNN to the Training set
        history = regressor.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=1)

        return regressor, history
   
    def __predict(self, test, regressor, window, sc):
        inputs = []
        for i in range(len(test)):
            inputs.append([test[i][1]])

        #print(inputs)
        inputs = sc.transform(inputs)
        #print("inputs",inputs)
        X_test = []
        y_test = []
        for i in range(window, len(inputs)):
            X_test.append(inputs[i-window:i, 0])
            y_test.append(inputs[i,0])

        #print("X_test",X_test)
        X_test = np.array(X_test)
        #print("X_test",X_test)
        X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
        #print("X_test",X_test)
        predicted_car_count = regressor.predict(X_test)
        predicted_car_count = sc.inverse_transform(predicted_car_count)

        score = regressor.evaluate(X_test, y_test)

        return predicted_car_count, score, regressor.metrics_names
  
    def __plot(self, training_set, predicted_car_count):
        import matplotlib.pyplot as plt

        data_and_predicted = training_set
        for i in range(len(predicted_car_count)):
            data_and_predicted.append(predicted_car_count[i])

        real_car_count = []
        for i in range(len(data_and_predicted)):
            real_car_count.append([self.numpyTraffic[i][1]])

        #add time on x-axis
        xtime = []
        for i in range(0,len(data_and_predicted)):
            xtime.append(self.numpyTraffic[i][0])#[dt.datetime.utcfromtimestamp(data[i][0]).strftime("%m/%d/%Y")])


        plt.plot(xtime, data_and_predicted, linestyle=":", color = 'blue', label = 'Predicted Stock Price')
        plt.plot(xtime, real_car_count,  color = 'red', label = 'Real Stock Price')
        plt.scatter(self.numpyTraffic[int(len(self.numpyTraffic)*0.8)-1][0],self.numpyTraffic[int(len(self.numpyTraffic)*0.8)-1][1], marker="|",color="black", s=5000)
        
        plt.ylabel('Opening Price', fontsize=14)
        #plt.xticks(np.arange(min(time), max(time)+1, 3.0), rotation=90)
        plt.xticks(rotation=90)
        #plt.xaxis.set_xticks(plt.get_xticks()[::2]) #skip every other x tick
        plt.title(self.file + " (" + str(len(self.numpyTraffic)) + " Readings)")
        plt.legend()
        #plt.grid(True)
        plt.show()
    
    def __plot_history(self, history):
        import matplotlib.pyplot as plt_history

        plt_history.plot(history['loss'], label = 'Loss')
        plt_history.plot(history['mae'], label = 'Mean Abs Error')
        plt_history.plot(history['accuracy'], label = 'Accuracy')

        #plt.plot(xtime, data_and_predicted, linestyle=":", color = 'blue', label = 'Predicted Stock Price')
        #plt.plot(xtime, real_car_count,  color = 'red', label = 'Real Stock Price')
        #plt.scatter(data[int(len(data)*0.8)-1][0],data[int(len(data)*0.8)-1][1], marker="|",color="black", s=5000)
        
        #plt.ylabel('Opening Price', fontsize=14)
        plt_history.xlabel('Epochs', fontsize=14)
        #plt.xticks(np.arange(min(time), max(time)+1, 3.0), rotation=90)
        #plt.xticks(rotation=90)
        #plt.xaxis.set_xticks(plt.get_xticks()[::2]) #skip every other x tick
        plt_history.title(self.file + " (" + str(len(data)) + " Readings)")
        plt_history.legend()
        #plt.grid(True)
        plt_history.show()
    
    def apply_LSTM(self, plot_prediction=True, plot_accuracy=True, col=100):
        epo = 10
        win = 20
        drop = 0.2 #inversly propotional to prediction accuracy
        #PRE_PROCESS DATA
        sc, X_train, y_train, test, training_set = self.__pre_process_data(
                                                            window=win,
                                                            col=col)
        
        print(sc)
        print(X_train)
        print(y_train)
        #TRAIN
        regressor = self.__build_regressor(X_train=X_train, y_train=y_train, 
                                units=50, dropout=drop )
        #FIT
        regressor, history = self.__fit(regressor, X_train=X_train, y_train=y_train, 
                                epochs=epo, batch_size=32)

                                
        #PREDICT
        predicted_car_count, score, metrics_names = self.__predict(test=test, 
                                regressor=regressor, 
                                window=win, sc=sc)
        print("Metric Names: ",metrics_names,"\nScore: ", score)

        #VISUALIZE
        # Visualising the results data with predicted stock prices
        if plot_prediction:
            self.__plot(training_set=training_set, predicted_car_count=predicted_car_count)
        if plot_accuracy:
            self.__plot_history(history.history)




def main():

    trafficObj = Traffic()
    #trafficObj.print_dataset(rows=rows)
    #trafficObj.add_factor(factor="fact_incident",reason="accident", year=2006, month=6)
    #trafficObj.add_factor(factor="fact_incident",reason="breakdown", year=2005, month=4)
    column = 3
    #trafficObj.print_dataset(rows=23, col=column)

    trafficObj.plot_traffic(col=column)
    trafficObj.apply_LSTM(plot_prediction=True, plot_accuracy=True, col=column)



# __name__
if __name__=="__main__":
    main()

```
